```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("timecpp.txt", "r")
times = file.read()
file.close()
times = times.split(", ")
del times[-1]
for i, val in enumerate(times):
    times[i] = float(val)

#times = times[:10]
times = np.asarray(times)
len(times)
```

```python
file = open("timecuda.txt", "r")
cuda = file.read()
file.close()
cuda = cuda.split(", ")
del cuda[-1]
for i, val in enumerate(cuda):
    cuda[i] = float(val)

#cuda = cuda[:10]
cuda = np.asarray(cuda)
len(cuda)
```

```python
x = np.linspace(0, 100000, 100)
len(x)
```

```python
plt.plot(x, times, label="C++", color="darkblue")
plt.plot(x, cuda, label="Cuda", color="darkred")

plt.xlabel("Array length")
plt.ylabel("time in [ms]")

plt.grid(linestyle="--", linewidth="0.8", color="grey")
plt.legend(loc="right")
#plt.savefig("performance.png")
plt.show()
```

```python
fig, ax1 = plt.subplots()
left, bottom, width, height = [0.12, 0.47, 0.5, 0.5]
ax2 = ax1.inset_axes([left, bottom, width, height],
                    xlim=(min(x), max(x)), ylim=(min(cuda) - 1, max(cuda) + 1),
                    xticklabels=[0, 2, 4, 6, 8, 10])

ax1.set_xlabel("array length")
ax1.set_ylabel("time in ms")

ax2.set_xlabel("AL * 10^-4")

ax1.plot(x, times, color="darkblue", label="C++")
ax1.plot(x, cuda, color="darkred", label="Cuda")

ax2.plot(x, cuda, color="darkred")

ax1.legend(loc="right")
ax1.indicate_inset_zoom(ax2)
ax1.grid(linestyle="--", linewidth=0.8, color="grey")
ax2.grid(linestyle="--", linewidth=0.8, color="grey")

plt.savefig("performance.png", dpi=1000)
plt.show()
```
