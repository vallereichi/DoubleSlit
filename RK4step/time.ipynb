```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("timecpp.txt", "r")
times = file.read()
file.close()
times = times.split(", ")
del times[-1]
for i, val in enumerate(times):
    times[i] = float(val)


times = np.asarray(times)
len(times)
```

```python
file = open("timecuda.txt", "r")
cuda = file.read()
file.close()
cuda = cuda.split(", ")
del cuda[-1]
for i, val in enumerate(cuda):
    cuda[i] = float(val)


cuda = np.asarray(cuda)
len(cuda)
```

```python
x = np.linspace(0, 100, 100)
len(x)
```

```python
plt.plot(x, times)
plt.plot(x, cuda)
plt.savefig("performance.png")
plt.show()
```
